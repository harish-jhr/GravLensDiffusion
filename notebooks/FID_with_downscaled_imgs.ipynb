```python
import os
import random
import torch
import numpy as np
import torchvision.transforms as transforms
from PIL import Image

# Paths
source_folder = "dir/Diffusion/data/Samples"
destination_folder = "dir/Diffusion/data/downscaled_100_pngs"
#downscale_size = (28)  = > Desired downscaled size

# Making the directory:
os.makedirs(destination_folder, exist_ok=True)

# Get all .npy files
npy_files = [f for f in os.listdir(source_folder) if f.endswith(".npy")]

# Check if there are enough files to sample
if len(npy_files) < 100:
    print("Not enough .npy files in the source folder!")
else:
    selected_files = random.sample(npy_files, 100)  # Randomly pick 100 .npy files

    for file in selected_files:
        # Load .npy file
        npy_path = os.path.join(source_folder, file)
        npy_data = np.load(npy_path)  # Shape assumed to be (1, H, W) for grayscale

        # Convert to PyTorch tensor
        tensor = torch.tensor(npy_data, dtype=torch.float32)  # Shape: (1, H, W)

        # Add batch dimension for interpolation: (B, C, H, W)
        tensor = tensor.unsqueeze(0)  # Shape: (1, 1, H, W)

        # Downscale using interpolate
        tensor_downscaled = torch.nn.functional.interpolate(
            tensor, size=(28), mode="bilinear", align_corners=False
        )

        # Remove batch dimension: (1, 28, 28) -> (28, 28)
        tensor_downscaled = tensor_downscaled.squeeze(0).squeeze(0)

        # Convert to PIL image
        transform_to_pil = transforms.ToPILImage()
        image = transform_to_pil(tensor_downscaled)

        # Save as PNG
        png_filename = os.path.splitext(file)[0] + ".png"  # Change extension
        image.save(os.path.join(destination_folder, png_filename))

    print("Successfully processed and saved 100 downscaled PNGs!")
```
